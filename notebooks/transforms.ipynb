```python
def get_delta() -> sm.SE3:
    import panda_py 
    import frankx
    import numpy as np
    from roboticstoolbox.models import Panda
    panda = Panda()

    q = panda.random_q()
    T_0p = sm.SE3(panda_py.fk(q), check = False)
    T_0f = sm.SE3(np.array(frankx.Kinematics.forward(q)).reshape(4,4, order = "F"))

    T_pf = T_0p.inv() * T_0f
    return T_pf

print(get_delta())
```
